```python
import seaborn as sns
import pandas as pd
import numpy as np

n = 50
summa = 0
f = open('1.txt')
arr = []
for i in range(n):
    x = float(f.readline()[:-1])
    summa += x
    arr.append(x)
average = round(summa/n, 2)
f.close()
```

```python
b = []
s = 0
for i in range(n):
    x = (arr[i] - average)**2
    b.append(x)
    s += x
    
sigma = (s/(n-1))**0.5
delta = round((2.01 * sigma) / (50)**0.5, 2)
print(average, delta)

tmin = min(arr)
tmax = max(arr)
print(tmin, tmax)
tdelta = round((tmax - tmin) / 5000, 5)

T = []
T.append(tmin)
for i in range(1, 5000):
    T.append(round(tdelta + T[i - 1], 5))
```

```python
from math import pi, exp
ro = []

for i in range(len(T)):
    x = (1/(sigma * (2 * pi) ** 0.5)) * exp(-((T[i] - average) ** 2) / (2 * sigma ** 2))  
    ro.append(round(x, 3))

rom = 1 / (sigma * (2 * pi) ** 0.5)


for i in range(len(T)):
    ro[i] = ro[i]
```

```python
sns.barplot(x=T, y=ro)
```

```python
import matplotlib.pyplot as plt
plt.hist(x=T, bins=50, density=True)
grid = np.linspace(0, 11, 11)
plt.plot(grid, color='red', label='Истинная плотность')

plt.legend()

plt.show()
```
